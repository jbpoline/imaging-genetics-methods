```python
import numpy as np
import scipy.stats as sst
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
%matplotlib inline
```

## Generate some values in an atlas from snp

```python
# generate some snp data
n_snp = 100
snp = np.random.normal(0,1,size=(n_snp,))
snp.shape
# plt.plot(snp)
sigma = 2.0
snp_filtered = ndi.filters.gaussian_filter1d(snp, sigma)
plt.plot(snp_filtered)

(maf1, maf2) = sst.scoreatpercentile(snp_filtered, (25,75))
snp_final = np.zeros_like(snp_filtered)
snp_final[snp_filtered>maf1] = 1
snp_final[snp_filtered>maf2] = 2
plt.plot(snp_final)
```

```python
## Generate some snp from a brain region atlas 
n_roi = 20
M = np.random.normal(0,1, size=(n_roi, n_snp))

roi_array = M.dot(snp_final)
```

```python
print(roi_array)
```

```python
# impose a correlation structure on the resulting atlas

roi_structure = np.random.normal(0,1,size=(n_roi, n_roi))
roi_structure[n_roi//2:-1, n_roi//2:-1] += 2
roi_structure[0:n_roi//2, 0:n_roi//2] -= 2
# plt.imshow(roi_structure)

M = np.random.normal(0,1, size=(n_roi, n_snp))
plt.plot(roi_structure.dot(M).dot(snp_final),'X')
```
